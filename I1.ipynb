```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/shootings.csv", header=None)

# Assign column names manually
columns = ['id', 'name','date','manner_of_death','armed','age','gender','race','city','state','signs_of_mental_illness','threat_level', 'flee', 'body_camera', 'other']
df.columns = columns

# Convert age to numeric and drop missing values
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df.dropna(subset=['age'], inplace=True)

# Plot modified histogram
plt.figure(figsize=(12, 6))
plt.hist(df['age'], bins=15, color='salmon', edgecolor='darkred')

# Add title and labels
plt.title('Age Distribution of Police Shooting Victims (2015â€“2020)', fontsize=14)
plt.xlabel('Age of Victim', fontsize=12)
plt.ylabel('Number of Victims', fontsize=12)

# Add grid lines
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

# Add mean line
mean_age = df['age'].mean()
plt.axvline(mean_age, color='black', linestyle='dashed', linewidth=2, label=f'Mean Age: {mean_age:.1f}')

# Add legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
